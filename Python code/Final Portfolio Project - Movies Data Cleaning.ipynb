#### Cleaning the 'Awards' column:
Presently 'Awards' just contains strings. The aim is to:
* Split the strings
* Retain only the number of wins/nomination 
* Convert the data into integers 
In this way the data can be used for analysis
  
The end product will be newly added columns that tell how many Oscars(or any other Award type mentioned) that a movie has won/been nominated for and its total wins and nominations

```python
#Checking Unique Values 
print( '\033[1m' + "Values contain either wins and/or nominations. In some cases the Award/Nomination Name is mentioned. The strings can be split by '.' or '&'" +  '\033[0m')

m_20['Awards'].unique()
```

```python
# Converting into string
m_20.Awards = m_20.Awards.astype('str')

#Creating a temporary DataFrame 'df' to contain single award entries from the 'Awards' column

#Splitting by '.' and stacking records 
df = pd.DataFrame((m_20.Awards.str.split('.', expand = True)).stack(), columns = ['awards_breakup'])
print( '\033[1m' + '\033[4m' + "Splitting by '.' and stacking records \n" + '\033[0m', df.tail(3))

#Splitting by '&' and stacking records 
df = pd.DataFrame(df['awards_breakup'].str.split('&', expand = True).stack(), columns = ['awards_breakup'])
print( '\033[1m' + '\033[4m' + "Splitting by '&' and stacking records \n" + '\033[0m', df.tail(3))


#Resetting index and removing irrelevant index columns creating by splitting
df.reset_index(inplace = True)
df.set_index('imdbID', inplace = True)
df.drop(['level_1','level_2'], inplace =True, axis = 1)

print( '\033[1m' + '\033[4m' + "df with single award entries \n" + '\033[0m', df.tail(6))


#Removing/ Filtering out  Null Values 
df.dropna(inplace = True)
#Removing 'nan'/ '' strings
df = df[(df['awards_breakup']!= 'nan') & (df['awards_breakup']!='')]

# Stripping off blank spaces
df.awards_breakup = df.awards_breakup.astype('str')
df.awards_breakup = df.awards_breakup.str.strip()
print( '\033[1m' + '\033[4m' + "df post data cleanup \n" + '\033[0m', df.tail(6))
```

```python
#Checking first 50 unique values in df
print('\033[1m'+ "Unique values show string containing 'win'/'wins' have unnamed awards; 'nomination'/'nominations' have unnamed nominations;\n'Won' have named awards & 'Nominated for' have named nominations" +  '\033[0m')
print('')
print(df.awards_breakup.unique()[:50])

# Checking if all cases have been counted for
print('')
print('')
print( '\033[1m' + "Records with:'Nominated for'" +  '\033[0m', df[df['awards_breakup'].str.contains('Nominated for')].shape)
print('\033[1m' + "Records with:'Won'" +  '\033[0m',df[df['awards_breakup'].str.contains('Won')].shape) 
print('\033[1m' + "Records with:'win'/'wins'" +  '\033[0m',df[df['awards_breakup'].str.contains('win')].shape)
print('\033[1m' + "Records with:'nomination'/'nominations'" +  '\033[0m',df[df['awards_breakup'].str.contains('nomination')].shape)
print('')
print('\033[1m' + "All records:" +  '\033[0m',df.shape)
print('\033[1m' + '\033[4m' +"Conclusion:" +  '\033[0m' + " All cases have been counted for. Sum of individual records add up to the shape of df")
```

```python
# Creating Columns for Named Awards, Named Nominations and Total Wins and Nominations and inputting their respective values
print( '\033[1m' + "Explanation: Named Awards and Named nominations values will be strings containing the name and the number of the awards/nominations. \n Presently, Total wins/ Nominations values  will be  values (integers) of Unnamed wins and nominations" + '\033[0m')

df['Named_Nominations'] = df.awards_breakup[df['awards_breakup'].str.contains('Nominated for')].astype('str').str.replace('Nominated for ','').astype('str')
df['Named_Awards'] = df.awards_breakup[df['awards_breakup'].str.contains('Won')].astype('str').str.replace('Won ','').astype('str')
df['Total_Wins'] = df.awards_breakup[df['awards_breakup'].str.contains('win')].astype('str').str.replace('Another ','').str.replace(' wins','').str.replace(' win','').astype('int')
df['Total_Nominations'] = df.awards_breakup[df['awards_breakup'].astype('str').str.contains('nomination')].str.replace('Another ','').str.replace(' nominations','').str.replace(' nomination','').astype('int')
print()
print(df[19:22])
print()
print( '\033[1m' + 'The New columns added have duplicated values.This is because the index of the orginal df did not have unique Index values.\n So for records having the same IMDB ID, the values in the new columns have been duplicated. \n We can fix this by grouping index values' + '\033[0m')
      
# Dropping 'awards_breakup' Column (Since each record here has unique values)
df.drop(['awards_breakup'], axis  = 1, inplace = True)

#Grouping Index values and finding max values.(Since all values in a group are the same in our case, duplicates get eliminated)
df = df.groupby('imdbID').max()
print()
print(df[df.index == 'tt0091983'])
print()
      

# Creating a column for each Named Nomination type such that the records contain Integer Values for each Nomination received
# Example: A new column df['Oscar_Nomination']  will show 1 if a record received 1 Oscar Nomination
      
# Creating a new column for each Named Nomination and inputting the respective number of nominations received. 
df.Named_Nominations = df.Named_Nominations.astype('str')
df['Golden_Globes_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Golden Globe')].str.replace(' Golden Globe','').str.replace('s','').astype('int')
df['Oscar_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Oscar')].str.replace(' Oscar','').str.replace('s','').astype('int')
df['Bafta_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('BAFTA')].str.replace(' BAFTA Film Award','').str.replace('s','').astype('int')
df['Primetime_Emmy_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Primetime')].str.replace(' Primetime Emmy','').str.replace('s','').astype('int')
print('\033[1m' + 'Post creating new columns for each Named Nomination Type'+ '\033[0m')
print(df[43:46])
#Simialarly, creating a column for each named award type and inputting the number of respective awards won
df.Named_Awards = df.Named_Awards.astype('str')
df['Golden_Globes_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Golden Globe')].str.replace(' Golden Globe','').str.replace('s','').astype('int')
df['Oscar_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Oscar')].str.replace(' Oscar','').str.replace('s','').astype('int')
df['Bafta_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('BAFTA')].str.replace(' BAFTA Film Award','').str.replace('s','').astype('int')
df['Primetime_Emmy_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Primetime')].str.replace(' Primetime Emmy','').str.replace('s','').astype('int')
print()
print('\033[1m' + 'Post creating new columns for each Named Award Type'+ '\033[0m')
print(df[50:51])

#Dropping 'Named_Nominations' and 'Named_Awards'
df.drop(['Named_Nominations','Named_Awards'], axis = 1, inplace = True)
      
#Adding Number of Named Awards and Nominations to Total Wins and Total Nominations
df.fillna(0, inplace = True)
df['Total_Nominations'] = df['Total_Nominations'] + df['Golden_Globes_Nomination'] + df['Oscar_Nomination'] + df['Bafta_Nomination'] + df['Primetime_Emmy_Nomination']
df['Total_Wins'] = df['Total_Wins'] + df['Golden_Globes_Award'] + df['Oscar_Award'] + df['Bafta_Award'] + df['Primetime_Emmy_Award']
pd.set_option('display.max.rows', 1000)
print(df.head(51))

##FIRST DRAFT FOR ABOVE CODE
# First draft performs at 3.66 s ± 407 ms per loop (mean ± std. dev. of 7 runs, 1 loop each)
#while above code performs at 2.17 ms ± 204 µs per loop (mean ± std. dev. of 7 runs, 100 loops each)

# %%timeit

# df.fillna(0, inplace = True)
# for x in df.index:
#     df.loc[x,'Total_Nominations'] = df.loc[x,'Total_Nominations'] + df.loc[x,'Golden_Globes_Nomination'] + df.loc[x,'Oscar_Nomination'] + df.loc[x,'Bafta_Nomination'] + df.loc[x,'Primetime_Emmy_Nomination']
#     df.loc[x,'Total_Wins'] = df.loc[x,'Total_Wins'] + df.loc[x,'Golden_Globes_Award'] + df.loc[x,'Oscar_Award'] + df.loc[x,'Bafta_Award'] + df.loc[x,'Primetime_Emmy_Award']


#Merging df with main dataset m_20. Dropping Awards column on m_20
m_20 = m_20.merge(df, on = 'imdbID', how = 'left').drop(columns = ['Awards'])
m_20
```

```python
# Creating Columns for Named Awards, Named Nominations and Total Wins and Nominations and inputting their respective values
print( '\033[1m' + "Explanation: Named Awards and Named nominations values will be strings containing the name and the number of the awards/nominations. \n Presently, Total wins/ Nominations values  will be  values (integers) of Unnamed wins and nominations" + '\033[0m')

df['Named_Nominations'] = df.awards_breakup[df['awards_breakup'].str.contains('Nominated for')].astype('str').str.replace('Nominated for ','').astype('str')
df['Named_Awards'] = df.awards_breakup[df['awards_breakup'].str.contains('Won')].astype('str').str.replace('Won ','').astype('str')
df['Total_Wins'] = df.awards_breakup[df['awards_breakup'].str.contains('win')].astype('str').str.replace('Another ','').str.replace(' wins','').str.replace(' win','').astype('int')
df['Total_Nominations'] = df.awards_breakup[df['awards_breakup'].astype('str').str.contains('nomination')].str.replace('Another ','').str.replace(' nominations','').str.replace(' nomination','').astype('int')
print()
print(df[19:22])
print()
print( '\033[1m' + 'The New columns added have duplicated values.This is because the index of the orginal df did not have unique Index values.\n So for records having the same IMDB ID, the values in the new columns have been duplicated. \n We can fix this by grouping index values' + '\033[0m')

# Dropping 'awards_breakup' Column (Since each record here has unique values)
df.drop(['awards_breakup'], axis  = 1, inplace = True)

#Grouping Index values and finding max values.(Since all values in a group are the same in our case, duplicates get eliminated)
df = df.groupby('imdbID').max()
print()
print(df[df.index == 'tt1776196'])
print()

# Creating a column for each Named Nomination type such that the records contain Integer Values for each Nomination received
# Example: A new column df['Oscar_Nomination']  will show 1 if a record received 1 Oscar Nomination
      
# Creating a new column for each Named Nomination and inputting the respective number of nominations received. 
df.Named_Nominations = df.Named_Nominations.astype('str')
df['Golden_Globes_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Golden Globe')].str.replace(' Golden Globe','').str.replace('s','').astype('int')
df['Oscar_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Oscar')].str.replace(' Oscar','').str.replace('s','').astype('int')
df['Bafta_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('BAFTA')].str.replace(' BAFTA Film Award','').str.replace('s','').astype('int')
df['Primetime_Emmy_Nomination'] = df['Named_Nominations'][df['Named_Nominations'].str.contains('Primetime')].str.replace(' Primetime Emmy','').str.replace('s','').astype('int')
print('\033[1m' + 'Post creating new columns for each Named Nomination Type'+ '\033[0m')
print(df[43:46])
#Simialarly, creating a column for each named award type and inputting the number of respective awards won
df.Named_Awards = df.Named_Awards.astype('str')
df['Golden_Globes_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Golden Globe')].str.replace(' Golden Globe','').str.replace('s','').astype('int')
df['Oscar_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Oscar')].str.replace(' Oscar','').str.replace('s','').astype('int')
df['Bafta_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('BAFTA')].str.replace(' BAFTA Film Award','').str.replace('s','').astype('int')
df['Primetime_Emmy_Award'] = df['Named_Awards'][df['Named_Awards'].str.contains('Primetime')].str.replace(' Primetime Emmy','').str.replace('s','').astype('int')
print()
print('\033[1m' + 'Post creating new columns for each Named Award Type'+ '\033[0m')
print(df[7:8])

#Dropping 'Named_Nominations' and 'Named_Awards'
df.drop(['Named_Nominations','Named_Awards'], axis = 1, inplace = True)
      
#Adding Number of Named Awards and Nominations to Total Wins and Total Nominations
df.fillna(0, inplace = True)
df['Total_Nominations'] = df['Total_Nominations'] + df['Golden_Globes_Nomination'] + df['Oscar_Nomination'] + df['Bafta_Nomination'] + df['Primetime_Emmy_Nomination']
df['Total_Wins'] = df['Total_Wins'] + df['Golden_Globes_Award'] + df['Oscar_Award'] + df['Bafta_Award'] + df['Primetime_Emmy_Award']
pd.set_option('display.max.rows', 1000)
print(df.head(51))

#Merging df with main dataset m_21. Dropping Awards column on m_21
m_21 = m_21.merge(df, on = 'imdbID', how = 'left').drop(columns = ['Awards'])
m_21
```

```python
#Converting hours & minutes format into minutes
m_21.Runtime = m_21.Runtime.astype('str')

#Creating dataframe with only 'h' and 'min' format runtimes. 
time = m_21[(m_21['Runtime'].str.contains('h')) & (m_21['Runtime'].str.contains('min'))]

#Creating a for loop. The 'Runtime' column in df time will be converted into integers and into minutes. Resultant time will
#then be assigned to the 'Runtime' column cells in df m_20 having the same index as df time
for x in time.index:
    m_21.loc[x, 'Runtime'] = int(time.loc[x,'Runtime'][0])*60 + int(time.loc[x,'Runtime'][4:6])
    
#Converting hours format into minutes    
m_21.Runtime = m_21.Runtime.astype('str')    
time = m_21[m_21['Runtime'].str.contains('h')]
for x in time.index:
    m_21.loc[x, 'Runtime'] = int(time.loc[x,'Runtime'][0])*60

#Converting minutes format values into integers
m_21.Runtime = m_21.Runtime.astype('str')
time = m_21[m_21['Runtime'].str.contains('min')]
for x in time.index:
    if m_21.loc[x, 'Runtime'] == '1.5 min':
        m_21 = m_21.drop(x)
    else:
        m_21.loc[x, 'Runtime'] = (int(time.loc[x,'Runtime'].replace('min','').replace(',','')))
        
del time 

m_21.head()
```
